```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

# Pooling

Pooling is downsample.

Pooling.png

```python
x = torch.rand([1, 16, 14, 14])
```

```python
layer = nn.MaxPool2d(2, stride = 2)   # nn.MaxPool2d(kernel_size, stride)
out = layer(x)
out.shape
```

```python
out = F.avg_pool2d(x, 2, stride = 2)   # same as previous code block
out.shape
```

# Upsample

upsample.png

```python
x = torch.rand([1, 16, 7, 7])
```

```python
out = F.interpolate(x, scale_factor = 2, mode = 'nearest')   # upsample function
out.shape
```

```python
out = F.interpolate(x, scale_factor = 3, mode = 'nearest')   # upsample function
out.shape
```

# ReLU

ReLU.png

```python
x = torch.rand([1, 16, 7, 7])
```

```python
layer = nn.ReLU(inplace = True)   # ReLU function
out = layer(x)
out.shape
```

```python
out = F.relu(x)   # same as previous code block
out.shape
```
